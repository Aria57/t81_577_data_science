```python
ale = ['ALE', 'IPA', 'PORTER', 'STOUT' ]
lager= ['LAGER', 'PILS', 'DUNK', 'BOCK']
df['category']=None

def re_label(beer_style):
    s=beer_style.replace(' ', '').upper()
    if any(sub in s for sub in ale):
        return 'ale'
    elif any(sub in s for sub in lager):
        return 'lager'
    else:
        return 'unknown'
```

```python
df['category'] = df['beer_style'].apply(re_label)
```
